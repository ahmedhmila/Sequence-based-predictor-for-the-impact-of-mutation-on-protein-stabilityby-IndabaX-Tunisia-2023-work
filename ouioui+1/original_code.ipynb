```python
wt_emb = torch.load("train/train_wt.pt")
mut_emb = torch.load("train/train_mut.pt")
df = pd.read_csv("train/train.csv")

# [Recommended] Split data into train and validation 
from sklearn.model_selection import train_test_split

# Split the data into train and validation (80/20 split)
train_df, val_df = train_test_split(df, test_size=0.2, random_state=42)

# Split the embeddings into train and validation
indices = train_df.index
wt_emb_train, mut_emb_train = wt_emb[indices], mut_emb[indices]

indices = val_df.index
wt_emb_val, mut_emb_val = wt_emb[indices], mut_emb[indices]

# Building the dataset class
class EmbeddingDataset(torch.utils.data.Dataset):
  def __init__(self,mut_pt, wt_pt, data_df):
    self.pt_mut = mut_pt
    self.pt_wt = wt_pt
    self.df = data_df
  
  def __len__(self):
      return self.pt_mut.shape[0]

  def __getitem__(self, index):
    o1=self.pt_mut[index,:]
    o2=self.pt_wt[index,:]
    if "ddg" in self.df:
      df_out=torch.Tensor([self.df["ddg"][index]])
    else:
      df_out=torch.Tensor([self.df["ID"][index]])
    return  self.pt_mut[index,:],self.pt_wt[index,:],df_out

# creating training and validation datasets 
train_dataset = EmbeddingDataset(wt_emb_train, mut_emb_train, train_df)
val_dataset = EmbeddingDataset(wt_emb_val, mut_emb_val, val_df)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from torch.utils.data import Dataset, DataLoader
# Define LSTM model with attention mechanism
class AttentionModel(torch.nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, layer_dim):
        super(AttentionModel, self).__init__()
        self.hidden_dim = hidden_dim
        self.layer_dim = layer_dim
        self.lstm = nn.LSTM(input_dim, hidden_dim, layer_dim, batch_first=True) 
        self.attention = nn.Linear(hidden_dim, 1)
        self.fc = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        h0 = torch.zeros(self.layer_dim, x.size(0), self.hidden_dim).requires_grad_()
        c0 = torch.zeros(self.layer_dim, x.size(0), self.hidden_dim).requires_grad_()
        out, (hn, cn) = self.lstm(x, (h0.detach(), c0.detach()))
        
        # Apply attention mechanism
        attn_weights = F.softmax(self.attention(out).squeeze(-1), dim=-1)
        out = torch.sum(out * attn_weights.unsqueeze(-1), dim=1)

        out = self.fc(out) 
        return out

# Create the model
model = AttentionModel(input_dim=1280, hidden_dim=512, output_dim=1, layer_dim=2)
```

```python
# Training parameters
epochs = 50
learning_rate = 0.001

# Define loss and optimizer
criterion = torch.nn.MSELoss()    # mean-squared error for regression
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)  # Adam optimizer

# Define data loaders
train_loader = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=32, shuffle=True)
val_loader = torch.utils.data.DataLoader(dataset=val_dataset, batch_size=32, shuffle=False)

# Training loop
for epoch in range(epochs):
    model.train()
    for i, (wt, mut, ddg) in enumerate(train_loader):
        optimizer.zero_grad()
        outputs = model(wt)
        loss = criterion(outputs, ddg)
        loss.backward()
        optimizer.step()
        
    # Validation
    model.eval()
    with torch.no_grad():
        total_val_loss = 0
        for i, (wt, mut, ddg) in enumerate(val_loader):
            outputs = model(wt)
            val_loss = criterion(outputs, ddg)
            total_val_loss += val_loss.item()
        
        print(f'Epoch: {epoch+1}, Train Loss: {loss.item():.4f}, Validation Loss: {total_val_loss/len(val_loader):.4f}')
```

```python
# preparing a dataloader for the training
train_dataloader = torch.utils.data.dataloader.DataLoader(
        train_dataset,
        batch_size=32,
        shuffle=False,
        num_workers=2,
    )
# [Recommended] Use Data validation loader too
```

```python
# Building a simple pytorch model
# A dummy model (2-param) that demonstrates the usage of the dataset

class StabilityModel(torch.nn.Module):
  def __init__(self):
    super(StabilityModel, self).__init__()
    self.lin = torch.nn.Linear(1,1)

  def forward(self, x, y):
    # run the forward pass
    # output should be the stability estimation [batch,estim]
    return self.lin(torch.mean(x-y,dim=1).reshape(-1,1)) 
```

```python
# Example of training script
device = torch.device("cuda")
model =  StabilityModel().to(device)
optimizer = torch.optim.Adadelta(model.parameters(), lr=0.0001)
criterion = torch.nn.MSELoss()
epoch_loss = 0
for i in range(1):
  epoch_loss = 0
  for batch_idx, (data_mut,data_wt , target) in tqdm(enumerate(train_dataloader)):
      # extract input from datallader
      x1 = data_wt.to(device)
      x2 = data_mut.to(device)
      y = target.to(device)
      # make prediction
      y_pred = model(x1,x2)
      # calculate loss and run optimizer
      loss = torch.sqrt(criterion(y, y_pred))
      loss.backward()
      optimizer.step()
      epoch_loss += loss
  print("epoch_",i," = ", epoch_loss/len(train_dataloader))
  # [Recommended] Save trained models to select best checkpoint for prediction (or add prediction in the epochs loop)
```
